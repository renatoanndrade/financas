```python
def p_delta(xf, xi):    
    delta_x = xf - xi
    porc_delta = (delta_x/xi)*100
    return porc_delta

def transf(brl, chz, xf):
    def calc_lucro(brl, chz, xf):    
        return chz*xf - brl
    l = calc_lucro(brl, chz, xf)
    return abs(l/(2-TX))

def od_compra(brl, chz, xf):
    t = transf(brl, chz, xf)   
    brl_f = brl - t
    chz_f = chz + (t/xf)*(1-TX)
    return brl_f, chz_f

def od_venda(brl, chz, xf):
    t = transf(brl, chz, xf)  
    brl_f = brl + t*(1-TX)
    chz_f = chz - (t/xf)
    return brl_f, chz_f

def s_venda(brl, chz, xf, xi):
    brl_t, chz_t = od_venda(brl, chz, xf)
    brl_f, chz_f = od_compra(brl_t, chz_t, xi)
    return brl_f, chz_f

def s_compra(brl, chz, xf, xi):
    brl_t, chz_t = od_compra(brl, chz, xf)
    brl_f, chz_f = od_venda(brl_t, chz_t, xi)
    return brl_f, chz_f

def pega_precos(moeda):
    klines = client.get_historical_klines(moeda, Client.KLINE_INTERVAL_1HOUR, "300 days ago UTC")
    precos = np.array([float(x[1]) for x in klines])
    return precos

def dinheiro_inicial():
    brl_i = M/(2+TX)
    chz_i = ((M-brl_i)*(1-TX))   
```

```python
from binance.client import Client
from secrets import api_secrets, api_key
from binance.enums import *
import matplotlib.pyplot as plt
import numpy as np
import plotly.express as px
from scipy import stats
from time import sleep
from IPython.display import clear_output
import winsound

client = Client(api_key, api_secrets)
print(client.get_account_status())

#pegando todas moedas
t = client.get_ticker()
todas_moedas = [x['symbol'] for x in t if x['symbol'][-3:] == 'USD']
M = 50_000
TX = 0.001
```

```python
empity = [] 
qty = []
def pega_porcs(moeda):
    MOEDA = moeda
    klines = client.get_historical_klines(MOEDA, Client.KLINE_INTERVAL_1HOUR, "300 days ago UTC")
    precos = np.array([float(x[1]) for x in klines])
    brl_i = M/(2+TX)
    chz_i = ((M-brl_i)*(1-TX))     
    if precos != empity:     
        xii = precos[1]
        chz_i = chz_i/xii
        xff = []
        var = 23
        xi = xii
        for x in range(2, len(precos)-1):             
            xf = precos[x]            
            dx_p = p_delta(xf, xi)            
            if dx_p > var:                
                qty.append(1)
                brl_i, chz_i = od_venda(brl_i, chz_i, xf) 
                xi = xf
            elif dx_p < -var:               
                qty.append(-1)
                brl_i, chz_i = od_compra(brl_i, chz_i, xf)  
                xi = xf   
        xf = xii*1.1
        C = brl_i + chz_i*xf*(1-TX)
        C2 = brl_i + chz_i*xii*(1-TX)
        porc = 100*(C- M)/M
        porc2 = 100*(C2- M)/M
    else: 
        porc = 'nan'
        porc2 = 'nan'
                
    return [porc, porc2, MOEDA, len(qty)]
```
